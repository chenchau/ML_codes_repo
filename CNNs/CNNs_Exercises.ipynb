## Optimization of a Single Neuron Using Gradient Descent - Numpy Example (Batch GD) in slide #31

```
import numpy as np 
import matplotlib.pyplot as plt 

def plot(X,y, axis=(1,2), loc = "upper left", theta=None):
  
  for i in np.unique(y):
    plt.scatter(X[np.squeeze(y)==i,axis[0]], X[np.squeeze(y)==i,axis[1]], s = 6)
  plt.xlabel(r'$x_{i,1}$')
  plt.ylabel(r'$x_{i,2}$')
  plt.legend(['Class {}'.format(i) for i in range(1,len(np.unique(y))+1)], loc=loc)
  if theta is not None: 
    theta = np.squeeze(theta)
    # the decision boundary for binary classification 
    y = lambda x: (0.5-theta[0]-theta[1]*x)/theta[2]
    axes = plt.gca()
    x0,x1 = axes.get_xlim()
    y0,y1 = axes.get_ylim()
    plt.plot([x0,x1],[y(x0), y(x1)],'k--')
    plt.axis([x0,x1,y0,y1])
```

**All mathematical notations**    


X (data matrix): $X \in R^{N X P}$        
 W (weight matrix):   
$W^k \in R^{P^k \times P^(k-1)}$,    k-th layer and (k-1)-th layer    
b: 
$b^k = [b_{1^k}^k, b_{2^k}^k, b_{3^k}^k,..., b_{P^k}^k]^\top$   
H()  hidden matrix will be the weighted sum of the previous activated output (plus the biases)    
$Y \in R^{N \times 1} = XW^T + ob^T$   


**W(weights matrix),  B(biases matrix), all in matrix form**   

in this section,we  do  (batch) gradient descents  

**Convergence:  a predetermined number of epochs have been completed**




```
X = np.array([[1., 1.],
             [9.4, 6.4],
             [2.5, 2.1],
             [8., 7.7],
             [0.5, 2.2],
             [7.9, 8.4],
             [7., 7.],
             [2.8, 0.8],
             [1.2, 3.],
             [7.8, 6.1]])   #shape: NxP (10x2)


Y = np.array([[1], [0], [1], [0], [1], [0], [0], [1], [1], [0]])# shape: Nx1
plt.figure(figsize=(5,5))
# plot the data,
plot(X, Y, axis=(0,1))

P_k_1 = X.shape[-1]   # P^(k-1) = P^(0) is the number of features of inputs
P_k = Y.shape[-1]   # P^(k) is the number of neurons of k-th layer, here is the last layer

num_samples = X.shape[0]
```

```
seed =1 
np.random.seed(seed)


# random initialization of single neuron parameters 
W = np.random.randn(P_k, P_k_1)   # should be 1xP(0)   [W11, W12, ..., W1p]
b = np.random.randn(P_k,1)  # this is b
o = np.ones([num_samples, 1])

# set up the hyperparameters for optimization (backpropagation)
alpha = 0.015 #learning rate 
max_iter = 200 # maximum number of iterations 

for i in range(max_iter):
  H = X@W.T + o@b.T  # XW^T (Nx1) + b
  Z = H  # using linear (identity mapping) activation 
  Y_hat = Z  
  
  # MSE loss  
  loss = np.sum((Y_hat-Y)**2)/num_samples   # Frobenius Norm
  
  # compute the gradients w.r.t. W(weight matrix)
  W_gradients = (2/num_samples)*(Y_hat.T - Y.T)@X
  
  b_gradients = (2/num_samples)*(Y_hat.T - Y.T)@o # as in the derivation
  
  # follow the negative gradients
  W = W - alpha*W_gradients   # update the weight matrix
  b = b - alpha*b_gradients   # update the biases matrix
  # train 
  Y_hat = X@W.T + o@b.T         # predicted "probabilities
  Y_hat[Y_hat>0.5] = 1    # if prob > 0.5 assign labels as ones 
  Y_hat[Y_hat<=0.5] = 0   # if prob <= 0.5 assign labels as zeros
  
  if (i+1)%10==0 or i==0:
    print("W gradients: ",np.squeeze(W_gradients))
    print("b gradients:",np.squeeze(b_gradients))
    print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} "
        .format(i+1, loss, np.mean(Y_hat==Y)))
```

```
# plotting training results  
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plot(X, Y, axis=(0,1))
plt.title('True labels for training')
plt.subplot(1,2,2)
#plot(X, Y_hat, axis=(0,1))
plot(X, Y_hat, axis=(0,1), theta=np.hstack((b[0].reshape(-1,1), W)))
plt.title('Predicted Labels for training')
```

## Optimization of a Single Neuron Using Gradient Descent - PyTorch Example (Batch GD) in slide #41

```
X = np.array([[1., 1.],
             [9.4, 6.4],
             [2.5, 2.1],
             [8., 7.7],
             [0.5, 2.2],
             [7.9, 8.4],
             [7., 7.],
             [2.8, 0.8],
             [1.2, 3.],
             [7.8, 6.1]])   #shape: NxP (10x2)


Y = np.array([[1], [0], [1], [0], [1], [0], [0], [1], [1], [0]])# shape: Nx1

# plot the data,
plot(X, Y, axis=(0,1))

P_k_1 = X.shape[-1]   # P^(k-1) = P^(0) is the number of features of inputs
P_k = Y.shape[-1]   # P^(k) is the number of neurons of k-th layer, here is the last layer
num_samples = X.shape[0]
```

```
import torch
import torch.nn.functional as F
seed =1 
np.random.seed(seed)
W = np.random.randn(P_k, P_k_1) # P_k_1=num_features
b = np.random.randn(P_k,1)  # this is b^T
o = np.ones([num_samples, 1])

# we have to convert all arrays to pytorch tensors 

X_tensor = torch.tensor(X).double()
y_tensor = torch.tensor(Y).double()
o_tensor = torch.tensor(o).double()

# since we are solving for theta, we need to use the flag "requires_grad"
W_tensor = torch.tensor(W,requires_grad=True).double()
b_tensor = torch.tensor(b,requires_grad=True).double()

alpha = 0.015 #learning rate 0.1
max_iter = 200 # maximum number of iterations 30
  

for i in range(max_iter): 
  y_hat = X_tensor@W_tensor.t() + o_tensor@b_tensor.t()

  # MSE loss
#   loss = torch.sum((y_tensor-y_hat)**2)/num_samples

  loss = F.mse_loss(y_hat, y_tensor)
  
  loss.backward()
  W_gradient = W_tensor.grad
  b_gradient= b_tensor.grad

  # updating the weights 
  with torch.no_grad():
    # optimize the objective function use the gradient descent
    W_tensor.data = W_tensor.data - alpha*W_gradient
    b_tensor.data = b_tensor.data - alpha*b_gradient

  
  # train 
  y_hat = X_tensor@W_tensor.t() + o_tensor@b_tensor.t()         # predicted "probabilities
  y_hat[y_hat>0.5] = 1
  y_hat[y_hat<=0.5] = 0
  
  if (i+1)%10==0 or i==0: 
  #validation 
    print("W gradients:",W_gradient.squeeze().numpy())
    print("b gradients:",b_gradient.squeeze().numpy())
    print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} "
        .format(i+1, loss.item(), torch.mean((y_hat==y_tensor).float())))
  W_tensor.grad.zero_() #clearing the gradients 
  b_tensor.grad.zero_()
  
```

```
# plotting training results  
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plot(X, Y, axis=(0,1))
plt.title('True labels for training')
plt.subplot(1,2,2)
plot(X, y_hat.detach().numpy(), axis=(0,1))
# plot(X, Y_hat, axis=(0,1), theta=np.hstack((B[0].reshape(-1,1), W)))
plt.title('Predicted Labels for training')
```

##optimization of a single neuron (with sigmoid activation and BCE loss)  using gradient descent - NumPy vs. PyTorch in slide # 55

**numpy implementation of a single neuron with sigmod activation and binary cross entropy loss**

```
seed =1 
np.random.seed(seed)


# random initialization of single neuron parameters 
W = np.random.randn(P_k, P_k_1)   # should be 1xP(0)   [W11, W12, ..., W1p]
b = np.random.randn(P_k,1)  # this is b
o = np.ones([num_samples, 1])

# set up the hyperparameters for optimization (backpropagation)
alpha = 0.06 #learning rate 
max_iter = 200 # maximum number of iterations 

for i in range(max_iter):
  H = X@W.T + o@b.T  # XW^T (Nx1) + b
  # using nonlinear (sigmoid) activation 
  Y_hat = 1. / (1. + np.exp(-H))

  # BCE loss
  loss = -1. * np.sum(Y * np.log(Y_hat) + (1-Y) * np.log(1-Y_hat)) / num_samples
  
  # compute the gradients w.r.t. W(weight matrix)
  W_gradients = (1./num_samples)*(Y_hat.T - Y.T)@X
  
  b_gradients = (1./num_samples)*(Y_hat.T - Y.T)@o # as in the derivation
  
  # follow the negative gradients
  W = W - alpha*W_gradients   # update the weight matrix
  b = b - alpha*b_gradients   # update the biases matrix
  # train 
  H = X@W.T + o@b.T  # XW^T (Nx1) + b
  Y_hat = 1. / (1. + np.exp(-H))         # predicted "probabilities
  Y_hat[Y_hat>0.5] = 1    # if prob > 0.5 assign labels as ones 
  Y_hat[Y_hat<=0.5] = 0   # if prob <= 0.5 assign labels as zeros
  
  if (i+1)%10==0 or i==0:
    print("W gradients: ",np.squeeze(W_gradients))
    print("b gradients:",np.squeeze(b_gradients))
    print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} "
        .format(i+1, loss, np.mean(Y_hat==Y)))
```

## Using Mini-Batch Gradient Descent - Numpy Example in slide #39

```
X = np.array([[1., 1.],
             [9.4, 6.4],
             [2.5, 2.1],
             [8., 7.7],
             [0.5, 2.2],
             [7.9, 8.4],
             [7., 7.],
             [2.8, 0.8],
             [1.2, 3.],
             [7.8, 6.1]])   #shape: NxP (10x2)


Y = np.array([[1], [0], [1], [0], [1], [0], [0], [1], [1], [0]])# shape: Nx1

# plot the data,
plot(X, Y, axis=(0,1))

P_k_1 = X.shape[-1]   # P^(k-1) = P^(0) is the number of features of inputs
P_k = Y.shape[-1]   # P^(k) is the number of neurons of k-th layer, here is the last layer

num_samples = X.shape[0]
```

```
seed =1 
np.random.seed(seed)


# random initialization of single neuron parameters 
W = np.random.randn(P_k, P_k_1)   # should be 1xP(0)   [W11, W12, ..., W1p]
b = np.random.randn(P_k,1)  # this is b
o = np.ones([num_samples, 1])

# set up the hyperparameters for optimization (backpropagation)
alpha = 0.015 #learning rate 
epochs = 200 # number of updates for all batches

# mini-batch gd 
batch_size = 5
# n_batches = int(num_samples/batch_size) # 2 batches per epoch


for epoch in range(epochs):
  indices = np.random.permutation(num_samples) # permute training samples
  X_permuted = X[indices]
  Y_permuted = Y[indices]
  for i in range(0,num_samples,batch_size):
    X_i = X_permuted[i:i+batch_size]    # mini-batch trainig samples
    Y_i = Y_permuted[i:i+batch_size]
    o_i = np.ones([batch_size, 1])
    # predicted     
    Y_hat_i= X_i@W.T + o_i@b.T  # XW^T (Nx1) + b

    # MSE loss  
    loss_i = np.sum((Y_hat_i-Y_i)**2)/batch_size   # notice the batch_size Frobenius Norm

    # compute the gradients w.r.t. W(weight matrix)
    W_gradients = (2/batch_size)*(Y_hat_i.T - Y_i.T)@X_i

    b_gradients = (2/batch_size)*(Y_hat_i.T - Y_i.T)@o_i # as in the derivation

    # follow the negative gradients
    W = W - alpha*W_gradients   # update the weight matrix
    b = b - alpha*b_gradients   # update the biases matrix
    # train 
    Y_hat = X@W.T + o@b.T         # predicted "probabilities
    Y_hat[Y_hat>0.5] = 1    # if prob > 0.5 assign labels as ones 
    Y_hat[Y_hat<=0.5] = 0   # if prob <= 0.5 assign labels as zeros
  
  if (epoch+1)%10==0 or epoch==0:
    print("W gradients: ",np.squeeze(W_gradients))
    print("b gradients:",np.squeeze(b_gradients))
    print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} "
        .format(epoch+1, np.sum((Y_hat-Y)**2)/num_samples, np.mean(Y_hat==Y)))
```

```
# plotting training results  
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plot(X, Y, axis=(0,1))
plt.title('True labels for training')
plt.subplot(1,2,2)
#plot(X, Y_hat, axis=(0,1))
plot(X, Y_hat, axis=(0,1), theta=np.hstack((b[0].reshape(-1,1), W)))
plt.title('Predicted Labels for training')
```

## Using Stochastic Gradient Descent - Numpy Example in slide #40

```
X = np.array([[1., 1.],
             [9.4, 6.4],
             [2.5, 2.1],
             [8., 7.7],
             [0.5, 2.2],
             [7.9, 8.4],
             [7., 7.],
             [2.8, 0.8],
             [1.2, 3.],
             [7.8, 6.1]])   #shape: NxP (10x2)


Y = np.array([[1], [0], [1], [0], [1], [0], [0], [1], [1], [0]])# shape: Nx1

# plot the data,
plot(X, Y, axis=(0,1))

P_k_1 = X.shape[-1]   # P^(k-1) = P^(0) is the number of features of inputs
P_k = Y.shape[-1]   # P^(k) is the number of neurons of k-th layer, here is the last layer

num_samples = X.shape[0]
```

```
seed =1 
np.random.seed(seed)


# random initialization of single neuron parameters 
W = np.random.randn(P_k, P_k_1)   # should be 1xP(0)   [W11, W12, ..., W1p]
b = np.random.randn(P_k,1)  # this is b
o = np.ones([num_samples, 1])

# set up the hyperparameters for optimization (backpropagation)
alpha = 0.0015 # NOTICE the learning rate needs smaller value, try original 0.015 instead
epochs = 200 # number of updates for all batches

rand_ind_list = np.arange(num_samples)
np.random.shuffle(rand_ind_list)

for epoch in range(epochs):
  for rand_ind in rand_ind_list:
    X_i = X[rand_ind,:].reshape(1,X.shape[1])  # only select one sample at a time
    Y_i = Y[rand_ind].reshape(1,1)
    o_i = np.ones([1, 1])
    # predicted     
    Y_hat_i= X_i@W.T + o_i@b.T  # XW^T (Nx1) + b

    # MSE loss  
    loss_i = np.sum((Y_hat_i-Y_i)**2)   # Frobenius Norm

    # compute the gradients w.r.t. W(weight matrix)
    W_gradients = (2/1)*(Y_hat_i.T - Y_i.T)@X_i

    b_gradients = (2/1)*(Y_hat_i.T - Y_i.T)@o_i # as in the derivation

    # follow the negative gradients
    W = W - alpha*W_gradients   # update the weight matrix
    b = b - alpha*b_gradients   # update the biases matrix
    # train 
    Y_hat = X@W.T + o@b.T         # predicted "probabilities
    Y_hat[Y_hat>0.5] = 1    # if prob > 0.5 assign labels as ones 
    Y_hat[Y_hat<=0.5] = 0   # if prob <= 0.5 assign labels as zeros
  
  if (epoch+1)%10==0 or epoch==0:
    print("W gradients: ",np.squeeze(W_gradients))
    print("b gradients:",np.squeeze(b_gradients))
    print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} "
        .format(epoch+1, np.sum((Y_hat-Y)**2)/num_samples, np.mean(Y_hat==Y)))
```

```
# plotting training results  
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plot(X, Y, axis=(0,1))
plt.title('True labels for training')
plt.subplot(1,2,2)
#plot(X, Y_hat, axis=(0,1))
plot(X, Y_hat, axis=(0,1), theta=np.hstack((b[0].reshape(-1,1), W)))
plt.title('Predicted Labels for training')
```

## Single Neuron Linear Classifier on  linearly-separable data (NumPy)


**1. The special plotting function**   
**2. The function to generate a simple dataset that is linearly seperable**



```
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split

def plot(X,y, axis=(1,2), loc = "upper left", theta=None):
  
  for i in np.unique(y):
    plt.scatter(X[np.squeeze(y)==i,axis[0]], X[np.squeeze(y)==i,axis[1]], s = 6)
  plt.xlabel(r'$x_{i,1}$')
  plt.ylabel(r'$x_{i,2}$')
  plt.legend(['Class {}'.format(i) for i in range(1,len(np.unique(y))+1)], loc=loc)
  if theta is not None: 
    theta = np.squeeze(theta)
    # the decision boundary for binary classification 
    y = lambda x: (0.5-theta[0]-theta[1]*x)/theta[2]
    axes = plt.gca()
    x0,x1 = axes.get_xlim()
    y0,y1 = axes.get_ylim()
    plt.plot([x0,x1],[y(x0), y(x1)],'k--')
    plt.axis([x0,x1,y0,y1])
    
    
def get_lin_sep_data(u1, u2, sig1, sig2, num_samples, test_size, add_bias=True, num_features=2):
    """get 2D linearly separable data"""
    # generate seperable normal distribution ~N(u1, sig1**2) and ~N(mu2, sig2**2)
    X1 = np.random.randn(int(num_samples/2), num_features)*sig1+u1
    X2 = np.random.randn(int(num_samples/2), num_features)*sig2+u2
    # all data
    X = np.concatenate((X1,X2), axis=0)
    if add_bias:
        # NumPy implementation: we add a bias term.
        # recall in regression slide #11: Xi^T = [1(bias term), Xi1, Xi2, ..., Xip]
        X = np.concatenate((np.ones((num_samples,1)),X), axis=1)
    # assign the ground truth labels for X1 as 0 and X2 as 1 
    y = np.concatenate((np.zeros((int(num_samples/2),1)),np.ones((int(num_samples/2),1))))
    # train-validation split 
    X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=test_size, random_state=seed)
    
    return X, X_train, X_validation, y, y_train, y_validation
```

**get two-dimensional linearly-separable data**   

 ***args:***   

*   ***num_samples***:   *The total number of samples to generate data  (default: 500)*    
*   ***test_size***: *The proportion of the dataset to include in the test split. (default 0.4)*
*   ***u1***: *mean of the first class. (default [0, 0])*    
*   ***u2***: *mean of the second class. (default [1.5, 2.5])*      
*   ***sig1***: *variance  of the first class. (default 0.5)*    
*   ***sig2***: *variance  of the second class. (default 0.4)*   



```
seed =1 
np.random.seed(seed)
num_features = 2
num_samples = 500
test_size = 0.4; 

## data parameters. Change those to get an easier/harder dataset
u1 = [0,0]     #mean of class 1 
u2 = [1.5,2.5] #mean of class 2 
sig1 = 0.5     #variance of class 1
sig2 = 0.4     #variance of class 3

# get 2D linearly separable data
X, X_train, X_validation, y, y_train, y_validation = get_lin_sep_data(u1, u2, sig1, sig2, num_samples, test_size, num_features)
```

**Visualizing the data** 

```
plt.figure(figsize=(15,4.5))
plt.subplot(1,3,1)
plot(X,y)
plt.title("all data")
plt.subplot(1,3,2)
plot(X_train, y_train)
plt.title("training data")
plt.subplot(1,3,3)
plot(X_validation, y_validation)
plt.title('validation data')
```

**Linear Classifier(w/o activations), Backpropagation**   

Let  $y$  be the ground truth and  $\hat{y}$   be the estimated values .   
Let $\theta$ be the parameters in the single Neuron and $\alpha$ be the learning rate.    


**We model the single Neuron as   $\hat{y} =X\theta$**    
***Assuming MSE loss***

$loss = (y-\hat{y})^2$

$\frac{\partial loss}{\partial \theta} = \frac{-2}{N}X^T(y-\hat{y})$

**Update the parameters using gradient descent**, 

$\theta_{t+1} = \theta_{t} - \alpha \frac{\partial loss}{\partial \theta} $  





 ***args:***   

*   ***alpha***:   *The learning rate  (default: 0.1)*    
*   ***max_iter***: *The maximum number of iterations. (default 30)*


```
np.random.seed(seed)

# random initilization of single neuron parameters 
theta = np.random.randn(num_features+1, 1)
# set up the hyperparameters for optimization (backpropagation)
alpha = 0.1 #learning rate 
max_iter = 30 # maximum number of iterations 

num_training_samples = X_train.shape[0]
for i in range(max_iter):
  y_train_hat = X_train@theta
  # MSE loss
  loss = ((y_train-y_train_hat)**2).mean()
  gradients = -2*X_train.T@((y_train-y_train_hat))/num_training_samples
  # follow the negative gradients
  theta = theta - alpha*gradients 
  # train 
  y_train_hat = X_train@theta         # predicted "probabilities
  y_train_hat[y_train_hat>0.5] = 1    # if prob > 0.5 assign labels as ones 
  y_train_hat[y_train_hat<=0.5] = 0   # if prob <= 0.5 assign labels as zeros
 
  #validation 
  y_validation_hat = X_validation@theta
  y_validation_hat[y_validation_hat>0.5] = 1
  y_validation_hat[y_validation_hat<=0.5] = 0

  print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} | Validation Accuracy = {:0.2f}"
        .format(i+1, loss, np.mean(y_train_hat==y_train),np.mean(y_validation_hat==y_validation)))

  
# plotting training results  
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plot(X_train, y_train)
plt.title('True labels for training')
plt.subplot(1,2,2)
plot(X_train, y_train_hat, theta=theta)
plt.title('Predicted Labels for training')



# plotting validation results  
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plot(X_validation, y_validation)
plt.title('True labels for validation')
plt.subplot(1,2,2)
plot(X_validation, y_validation_hat, theta=theta)
plt.title('Predicted Labels for validation')
```

**The function to generate a linearly-nonseparable data  (a cloud in a ring)** 

```
import numpy as np 
from sklearn.model_selection import train_test_split



def get_non_lin_data(u1, sig1, radius2, num_samples, test_size, add_bias=True, num_features=2):
    """get 2D non-linear data"""
    # a cloud in a ring 
    X1 = np.random.randn(int(num_samples/2),num_features)*sig1+u1
    X2 = np.random.rand(int(num_samples/2),num_features)*2*radius2-radius2
    X2[:,1] = np.random.choice((-1, 1), size=int(num_samples/2))*np.sqrt(radius2**2-X2[:,0]**2)+ np.random.randn(int(num_samples/2))*radius2/10
    # concatenate the samples in the real challenge
    X = np.concatenate((X1,X2), axis=0)
    if add_bias:
      # here we add a bias term.
      X = np.concatenate((np.ones((int(num_samples/2)*2,1)),X), axis=1)
    y = np.concatenate((np.zeros((int(num_samples/2),1)),np.ones((int(num_samples/2),1))))
    X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=test_size, random_state=seed)
    return X, X_train, X_validation, y, y_train, y_validation
```

**non-linear Classifier(sigmoid activations), Backpropagation**   

Let  $y$  be the ground truth and  $\hat{y}$   be the estimated values .   
Let $\theta$ be the parameters in the single Neuron and $\alpha$ be the learning rate.    

**non-linearity (sigmoid activation)** :   

$\sigma(x) = \frac{1}{1+e^{-x}}$   
**We model the single neuron non-linear classifier as   $\hat{y} = \sigma(X\theta) = \frac{1}{1+e^{-X\theta}}$**    



***Assuming MSE loss***

$loss = (y-\hat{y})^2$    

step1: $\frac{\partial loss}{\partial \hat{y}} =\frac{-2}{N}(y-\hat{y}) $    
step2 : $\frac{\partial \hat{y}}{\partial (X\theta)} = \frac{e^{-X\theta}}{(1+e^{-X\theta})^2} = (1-\sigma{(X\theta)}) \sigma{(X\theta)}$     
step3: $\frac{\partial X\theta}{\partial \theta } = X^T$   
**Using chain rule**: 
$\Delta = \frac{\partial loss}{\partial \theta} = \frac{-2}{N}X^T(1-\sigma{(X\theta)}) \sigma{(X\theta)}(y-\hat{y})$

Then, 

$\theta_{t+1} = \theta_{t} - \alpha \Delta $


 ***args:***   

*   ***alpha***:   *The learning rate  (default: 0.1)*    
*   ***max_iter***: *The maximum number of iterations. (default 30)*


```
np.random.seed(seed)

# random initilization of single neuron parameters
theta = np.random.rand(num_features+1,1)

alpha = 0.1 #learning rate 0.1
max_iter = 30 # maximum number of iterations 30
num_training_samples = X_train.shape[0]


for i in range(max_iter):
  # sigmoid activation
  y_train_hat =  1. / (1 + np.exp(-X_train@theta))
  # MSE loss
  loss = ((y_train-y_train_hat)**2).mean()
#   gradients = -2*X_train.T@((y_train-y_train_hat))/num_training_samples
  dloss_dy_train_hat = -2 * (y_train - y_train_hat) / num_training_samples   # derivative of loss w.r.t. \hat{y}
  dy_train_hat_dXtheta = (1 - y_train_hat) * y_train_hat  # derivative of \hat{y} w.r.t. X\theta
  pXtheta_ptheta = X_train   # partial derivative of X\theta w.r.t. \theta
  # compute the gradients by chain rule
  gradients = pXtheta_ptheta.T@(dy_train_hat_dXtheta * dloss_dy_train_hat)
  # follow the negative gradients
  theta = theta - alpha*gradients
  
  y_train_hat[y_train_hat>0.5] = 1
  y_train_hat[y_train_hat<=0.5] = 0
  
  #validation 
  y_validation_hat = X_validation@theta
  y_validation_hat[y_validation_hat>0.5] = 1
  y_validation_hat[y_validation_hat<=0.5] = 0
  print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} | Validation Accuracy = {:0.2f}".format(i+1, loss, np.mean(y_train_hat==y_train),np.mean(y_validation_hat==y_validation)))
```

```
# plotting training results  
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plot(X_train, y_train)
plt.title('True labels for training')
plt.subplot(1,2,2)
plot(X_train, y_train_hat, theta=None)
plt.title('Predicted Labels for training')
# plt.savefig('single_neuron_nonlin_classfier_nonlin_data_train.png')

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plot(X_validation,y_validation)
plt.title('True Labels for validation')
plt.subplot(1,2,2)
plot(X_validation,y_validation_hat,theta=None)
plt.title('Predicted Labels for validation')
# plt.savefig('single_neuron_nonlin_classfier_nonlin_data_val.png')
```

##Single neuron non-linear (sigmoid) classifier with  BCE loss on the linearly-nonseparable data (NumPy)  
**Binary Cross Entropy Loss (BCE loss)**     



**get two-dimensional linearly-separable data (a cloud in a ring)**   

 ***args:***   

*   ***num_samples***:   *The total number of samples to generate data  (default: 500)*    
*   ***test_size***: *The proportion of the dataset to include in the test split. (default 0.5)*

*   ***u1***: *mean of the first class. (default [0, 0])*    
*   ***sig1***: *variance of the first class. (default 0.1)*   
*   ***radius2***: *radius  of the second class. (default 0.7)*

```
seed =100
np.random.seed(seed)
num_features = 2
num_samples = 500
test_size = 0.5; 

# data parameters, change them to get a simpler/harder dataset
u1 = [0,0]      #mean of class 1 
sig1 = 0.1      #variance of class 1
radius2 = 0.7;

X, X_train, X_validation, y, y_train, y_validation = get_non_lin_data(u1, sig1, radius2, num_samples, test_size)

plt.figure(figsize=(20, 6))
plt.subplot(1,3,1)
plot(X,y)
plt.title("all data")
plt.subplot(1,3,2)
plot(X_train, y_train)
plt.title("training data")
plt.subplot(1,3,3)
plot(X_validation, y_validation)
plt.title('validation data')
```

**non-linear Classifier(sigmoid activations), Backpropagation**   

Let  $y$  be the ground truth and  $\hat{y}$   be the estimated values .   
Let $\theta$ be the parameters in the single Neuron and $\alpha$ be the learning rate.    

**non-linearity (sigmoid activation)** :   

$\sigma(x) = \frac{1}{1+e^{-x}}$   
**We model the single neuron non-linear classifier as   $\hat{y} = \sigma(X\theta) = \frac{1}{1+e^{-X\theta}}$**    



***Assuming Binary Cross Entropy Loss (BCE loss)***

$BCE loss = -\frac{1}{N}\sum_{i=1}^{N}y_i\log(\hat{y_i})+(1-y_i)\log{(1-\hat{y_i})}$    

step1: $\frac{\partial loss}{\partial \hat{y}} =\frac{-1}{N}(\frac{y_i}{\hat{y_i}}-\frac{1-y_i}{1-\hat{y}})   $    
step2 : $\frac{\partial \hat{y}}{\partial (X\theta)} = \frac{e^{-X\theta}}{(1+e^{-X\theta})^2} = (1-\sigma{(X\theta)}) \sigma{(X\theta)}$     
step3: $\frac{\partial X\theta}{\partial \theta } = X^T$   
**Using chain rule**: 
$\Delta = \frac{\partial loss}{\partial \theta} = \frac{-1}{N}X^T(1-\sigma{(X\theta)}) \sigma{(X\theta)}(\frac{y_i}{\hat{y_i}}-\frac{1-y_i}{1-\hat{y}}) $

Then, 

$\theta_{t+1} = \theta_{t} - \alpha \Delta $


 ***args:***   

*   ***alpha***:   *The learning rate  (default:5.9)*    
*   ***max_iter***: *The maximum number of iterations. (default 60)*


```
np.random.seed(seed)

# random initilization of single neuron parameters
theta = np.random.rand(num_features+1,1)

alpha = 0.9 #learning rate 0.1
max_iter = 60 # maximum number of iterations 30
num_training_samples = X_train.shape[0]


for i in range(max_iter):
  # sigmoid activation
  y_train_hat =  1. / (1 + np.exp(-X_train@theta))
  # BCE loss
  loss = -1. * np.sum(y_train * np.log(y_train_hat) + (1-y_train) * np.log(1-y_train_hat)) / num_training_samples
  dloss_dy_train_hat = -1.*((y_train*1./y_train_hat) - (1-y_train)*1./(1-y_train_hat))/ num_training_samples   # derivative of loss w.r.t. \hat{y}
  
  dy_train_hat_dXtheta = (1 - y_train_hat) * y_train_hat  # derivative of \hat{y} w.r.t. X\theta
  pXtheta_ptheta = X_train   # partial derivative of X\theta w.r.t. \theta
 
  # compute the gradients by chain rule
  gradients = pXtheta_ptheta.T@(dy_train_hat_dXtheta * dloss_dy_train_hat)
  # follow the negative gradients
  theta = theta - alpha*gradients
  
  y_train_hat[y_train_hat>0.5] = 1
  y_train_hat[y_train_hat<=0.5] = 0
  
  #validation 
  y_validation_hat = 1. / (1 + np.exp(-X_validation@theta))
  y_validation_hat[y_validation_hat>0.5] = 1
  y_validation_hat[y_validation_hat<=0.5] = 0
  print("Iteration:{} | Loss = {:0.4f}  | Training Accuracy = {:0.2f} | Validation Accuracy = {:0.2f}".format(i+1, loss, np.mean(y_train_hat==y_train),np.mean(y_validation_hat==y_validation)))
```

```
import numpy as np 
from sklearn.model_selection import train_test_split



def get_non_lin_data(u1, sig1, radius2, num_samples, test_size, add_bias=True, num_features=2):
    """get 2D non-linear data"""
    # a cloud in a ring 
    X1 = np.random.randn(int(num_samples/2),num_features)*sig1+u1
    X2 = np.random.rand(int(num_samples/2),num_features)*2*radius2-radius2
    X2[:,1] = np.random.choice((-1, 1), size=int(num_samples/2))*np.sqrt(radius2**2-X2[:,0]**2)+ np.random.randn(int(num_samples/2))*radius2/10
    # concatenate the samples in the real challenge
    X = np.concatenate((X1,X2), axis=0)
    if add_bias:
      # here we add a bias term.
      X = np.concatenate((np.ones((int(num_samples/2)*2,1)),X), axis=1)
    y = np.concatenate((np.zeros((int(num_samples/2),1)),np.ones((int(num_samples/2),1))))
    X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=test_size, random_state=seed)
    return X, X_train, X_validation, y, y_train, y_validation
```


### PyTorch Module API
PyTorch provides the `nn.Module` API for you to define arbitrary network architectures, while tracking every learnable parameters for you. PyTorch also provides the `torch.optim` package that implements all the common optimizers, such as SGD, RMSProp, Adagrad, and Adam. You can refer to the [doc](http://pytorch.org/docs/master/optim.html) for the exact specifications of each optimizer.

To use the Module API, follow the steps below:

1. Subclass `nn.Module`.  

2. In the constructor `__init__()`, define all the layers you need as class attributes. Layer objects like `nn.Linear` and `nn.Conv2d` are themselves `nn.Module` subclasses and contain learnable parameters, so that you don't have to instantiate the raw tensors yourself. `nn.Module` will track these internal parameters for you.
```
from sklearn.model_selection import train_test_split


DATA_FLAG = "iris"
test_size = 0.25

X, y = datasets[DATA_FLAG]['X'], datasets[DATA_FLAG]['y'] 
                                                     

# train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y,                
                                          test_size = test_size, random_state=1)

# change the values below to have wider/narrower hidden layers
# or add values to stack more hidden layers, e.g. [50, 50, 25]
hidden_layer_sizes=[60, 40, 20]
activation = "relu"
solver='sgd'
learning_rate_init = 0.04
max_iter=500
batch_size= 16

alpha = 1e-4
tol=1e-4
```

```
def load_data_tensor(X, y, norm=True, shuffle=False):
    """load samples and labels in batches"""
    if norm:
       X /= X.max(axis=0)
    if shuffle:
       X = np.random.permutation(X)
       y = np.random.permutation(y)
    # convert the numpy array into torch tensor
    X_tensor = torch.FloatTensor(X)     # N x P, N: number of samples, P: number of features
    y_tensor = torch.from_numpy(y)     # N,
    
    return X_tensor, y_tensor
  
  
X_train_tensor, y_train_tensor = load_data_tensor(X_train, y_train, norm=True, shuffle=True)
```

```
import torch
import torch.nn as nn

# Give your network class an intuitive name like `MLPClassifier`.
class MLPClassifier(nn.Module):
    # in constructor define all the layers you need as class attributes
    def __init__(self, input_size, hidden_layer_sizes, num_classes):
        super().__init__()
        
        layers = [] 
        # assign layer objects to class attributes
        layers += [nn.Linear(input_size, hidden_layer_sizes[0]), nn.ReLU(inplace=True)]
        for i in range(0, len(hidden_layer_sizes)-1):
            if activation == "relu":
                layers += [nn.Linear(hidden_layer_sizes[i], hidden_layer_sizes[i+1]), nn.ReLU(inplace=True)]
            else:
                layers += [nn.Linear(hidden_layer_sizes[i], hidden_layer_sizes[i+1])]
        # the last hidden layer output the prediction for each category
        layers += [nn.Linear(hidden_layer_sizes[-1], num_classes)]
        
        self.layers =  nn.Sequential(*layers)

        
    def forward(self, x):
        logits = self.layers(x)
        return logits

```

```
from torch.autograd import Variable
import torch.optim as optim

# define the input and output dimension from data
input_dim = X_train.shape[-1]
output_dim = len(np.unique(y_train))

## define the MLP classifier
model = MLPClassifier(input_size=input_dim, 
                     hidden_layer_sizes=hidden_layer_sizes,
                    num_classes = output_dim)

# model summary
print(model)

# defeine cross entropy loss for multi-class classification 
loss_fun = torch.nn.CrossEntropyLoss()

# define the optimizer
if solver =='sgd':
  # SGD optimizer
  optimizer= torch.optim.SGD(model.parameters(), lr=learning_rate_init)



# training
for epoch in range(max_iter):
#     print("Epoch: "+str(epoch))
    for i in range(0, X_train_tensor.size()[0], batch_size):
#         print("batch: "+str(i))
        
        X_batch, y_batch = X_train_tensor[i:i+batch_size], y_train_tensor[i:i+batch_size]
        X_var = Variable(X_batch)
        y_var = Variable(y_batch)
    
        y_pred_var = model(X_var)
        # compute the loss
        loss = loss_fun(y_pred_var, y_var)
        

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
print(epoch, loss.item())

# def test_TwoLayerFC():
#     input_size = 50
#     x = torch.zeros((64, input_size), dtype=dtype)  # minibatch size 64, feature dimension 50
#     model = TwoLayerFC(input_size, 42, 10)
#     scores = model(x)
#     print(scores.size())  # you should see [64, 10]
# test_TwoLayerFC()
```
